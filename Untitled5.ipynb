```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

def load_data(file_path):
    data = pd.read_csv(file_path)
    data.columns = data.columns.str.lower()
    return data

def preprocess_data(data):
    features = data[['customerid', 'gender', 'age', 'annual income (k$)', 'spending score (1-100)']].copy()
    features['gender'] = features['gender'].map({'Male': 0, 'Female': 1})
    features[['age', 'annual income (k$)', 'spending score (1-100)']] = features[[
        'age', 'annual income (k$)', 'spending score (1-100)'
    ]].apply(lambda x: x.fillna(x.median()))

    if features['gender'].isna().any() or not np.isfinite(features['gender']).all():
        features['gender'] = features['gender'].fillna(-1).astype('int')
    else:
        features['gender'] = features['gender'].astype('int')

    return features

def find_optimal_k(data, max_k=10):
    wcss = []
    scaled_features = StandardScaler().fit_transform(data.drop('customerid', axis=1))

    for i in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
        kmeans.fit(scaled_features)
        wcss.append(kmeans.inertia_)

    return wcss

def visualize_elbow_method(wcss):
    plt.plot(range(1, len(wcss) + 1), wcss, marker='o')
    plt.title('Elbow Method for Optimal k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
    plt.grid(True)
    plt.show()

def visualize_clusters(data, feature_names):
    if len(feature_names) >= 2:
        plt.scatter(data[feature_names[0]], data[feature_names[1]], c=data['Cluster'], cmap='viridis', alpha=0.8)
        plt.title(f'K-means Clustering of Customers - {feature_names[0]} vs {feature_names[1]}')
        plt.xlabel(feature_names[0])
        plt.ylabel(feature_names[1])
        plt.show()
    else:
        print("Invalid number of features for visualization.")

def main():
    file_path = '/content/drive/MyDrive/purchase/Mall_Customers.csv'
    data = load_data(file_path)
    features = preprocess_data(data)

    # Determine the optimal number of clusters
    wcss = find_optimal_k(features)
    visualize_elbow_method(wcss)

    # Apply K-means clustering with the chosen k
    optimal_k = 5  # Adjust based on the graph
    scaled_features = StandardScaler().fit_transform(features.drop('customerid', axis=1))
    kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    clusters = kmeans.fit_predict(scaled_features)

    # Add the cluster labels to the original dataframe
    data['Cluster'] = clusters

    #cluster graph
    print('Cluster Graph is :')

    # Visualize the clusters in 2D space
    visualize_clusters(data, ['annual income (k$)', 'spending score (1-100)'])

    # Display Customer IDs and Assigned Clusters
    customer_clusters = data[['customerid', 'Cluster']]
    print("\nCustomer IDs and Assigned Clusters:")
    print(customer_clusters)

if __name__ == "__main__":
    main()

```
